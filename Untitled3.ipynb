```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# Lecture du fichier CSV
df = pd.read_csv('/content/water_potability.csv')

# Description des donnes
print(df.info())
print(df.head())
print(df.tail())

# Affichage du nombre de lignes dans le dataframe
num_records = len(df)
print(f"Number of Records: {num_records}")

# Affichage des types de donnes de chaque colonne
print("\nColumn names and data types:")
print(df.dtypes)

# Affichage des statistiques descriptives
print(df.describe())

# Vrification des valeurs manquantes
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Suppression des valeurs manquantes
df_imputed = df.dropna()
print('Data after dropping missing values:')
print(df_imputed.head())

# Visualisation des distributions avec des boxplots
fig, axs = plt.subplots(5, 2, figsize=(15, 20))
axs = axs.flatten()
for i, col in enumerate(df.columns[:-1]):
    sns.boxplot(x=df[col], ax=axs[i])
    axs[i].set_title(col)
plt.tight_layout()
plt.show()

# Visualisation des distributions avec des histogrammes
fig, axs = plt.subplots(5, 2, figsize=(15, 20))
Potable = df[df["Potability"] == 0]
No = df[df["Potability"] == 1]

for ax, col in zip(axs.flatten(), df.columns[:-1]):
    ax.hist(Potable[col], bins=20, label="Potable", alpha=0.7)
    ax.hist(No[col], bins=20, label="No", alpha=0.7)
    ax.set_title(col)

axs[4, 1].axis('off')  # Masquer le dernier sous-graphique inutile
plt.tight_layout()
plt.show()

# Visualisation des relations avec des nuages de points
print("Scatter Plots:\n")
features = ['ph', 'Sulfate', 'Trihalomethanes', 'Chloramines', 'Conductivity', 'Hardness', 'Turbidity', 'Organic_carbon', 'Solids']
for i in range(0, len(features)-1, 2):
    df_imputed.plot.scatter(x=features[i], y=features[i+1])
    plt.show()

# Matrice de corrlation
print(df.iloc[:, 0:9].corr())

# Gestion des outliers pour chaque caractristique
features = ['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']
for feature in features:
    Q1 = df_imputed[feature].quantile(0.25)
    Q3 = df_imputed[feature].quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5
    outliers = df_imputed[(df_imputed[feature] < Q1 - threshold * IQR) | (df_imputed[feature] > Q3 + threshold * IQR)]
    df_imputed.loc[outliers.index, feature] = df[feature].mean()

# Calcul des variances
print("Variance for each feature:")
for feature in features:
    variance = df_imputed[feature].var()
    print(f"{feature}: {variance}")

# Division des donnes en ensemble d'entranement et de test
x = df_imputed.iloc[:, :-1]
y = df_imputed["Potability"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(x_train.shape, y_train.shape)

# Cration et entranement du modle Decision Tree
treeClassifier = DecisionTreeClassifier()
treeClassifier.fit(x_train, y_train)

y_test_pred = treeClassifier.predict(x_test)
# Affichage des proprits de l'arbre
print("Number of leaves:", treeClassifier.get_n_leaves())
print("Tree depth:", treeClassifier.get_depth())
plt.figure(figsize=(20, 10))
plot_tree(treeClassifier, filled=True)
plt.show()
# Calculate the accuracy of the model on the test set
accuracy_test = accuracy_score(y_test, y_test_pred)

print("Accuracy on the test set of decesion tree:", accuracy_test)
# Initialize and train a logistic regression model
model = LogisticRegression()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Calculate the accuracy of the model on the test set
accuracy_test = accuracy_score(y_test, y_pred)
print("Accuracy on the test set of LogisticRegression :", accuracy_test)
# Initialize and train a SVM model
rs = 123
model = SVC()
model.fit(x_train, y_train.values.ravel())
preds = model.predict(x_test)
# Calculate the accuracy of the model on the test set
accuracy_test = accuracy_score(y_test, preds)
print("Accuracy on the test set of SVM :", accuracy_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
x = df_imputed.iloc[:, :-1]
y = df_imputed["Potability"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(x_train.shape, y_train.shape)
# Decision Tree
param_grid_tree = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_tree = GridSearchCV(DecisionTreeClassifier(), param_grid_tree, cv=5, scoring='accuracy')
grid_tree.fit(x_train, y_train)
best_tree = grid_tree.best_estimator_
print("Best parameters for Decision Tree:", grid_tree.best_params_)
print("Best Decision Tree Accuracy:", grid_tree.best_score_)

# Logistic Regression
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'saga']
}
grid_lr = GridSearchCV(LogisticRegression(max_iter=10000), param_grid_lr, cv=5, scoring='accuracy')
grid_lr.fit(x_train, y_train)
best_lr = grid_lr.best_estimator_
print("Best parameters for Logistic Regression:", grid_lr.best_params_)
print("Best Logistic Regression Accuracy:", grid_lr.best_score_)

```

```python
# SVM
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
grid_svm = GridSearchCV(SVC(), param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(x_train, y_train)
best_svm = grid_svm.best_estimator_
print("Best parameters for SVM:", grid_svm.best_params_)
print("Best SVM Accuracy:", grid_svm.best_score_)
from sklearn.ensemble import VotingClassifier

voting_clf = VotingClassifier(estimators=[
    ('dt', best_tree),
    ('lr', best_lr),
    ('svm', best_svm)
], voting='hard')

voting_clf.fit(x_train, y_train)
voting_pred = voting_clf.predict(x_test)
voting_accuracy = accuracy_score(y_test, voting_pred)
print("Voting Classifier Accuracy:", voting_accuracy)
```
