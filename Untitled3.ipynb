```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# Lecture du fichier CSV
df = pd.read_csv('/content/water_potability.csv')

# Description des données
print(df.info())
print(df.head())
print(df.tail())

# Affichage du nombre de lignes dans le dataframe
num_records = len(df)
print(f"Number of Records: {num_records}")

# Affichage des types de données de chaque colonne
print("\nColumn names and data types:")
print(df.dtypes)

# Affichage des statistiques descriptives
print(df.describe())

# Vérification des valeurs manquantes
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Suppression des valeurs manquantes
df_imputed = df.dropna()
print('Data after dropping missing values:')
print(df_imputed.head())

# Visualisation des distributions avec des boxplots
fig, axs = plt.subplots(5, 2, figsize=(15, 20))
axs = axs.flatten()
for i, col in enumerate(df.columns[:-1]):
    sns.boxplot(x=df[col], ax=axs[i])
    axs[i].set_title(col)
plt.tight_layout()
plt.show()

# Visualisation des distributions avec des histogrammes
fig, axs = plt.subplots(5, 2, figsize=(15, 20))
Potable = df[df["Potability"] == 0]
No = df[df["Potability"] == 1]

for ax, col in zip(axs.flatten(), df.columns[:-1]):
    ax.hist(Potable[col], bins=20, label="Potable", alpha=0.7)
    ax.hist(No[col], bins=20, label="No", alpha=0.7)
    ax.set_title(col)

axs[4, 1].axis('off')  # Masquer le dernier sous-graphique inutile
plt.tight_layout()
plt.show()

# Visualisation des relations avec des nuages de points
print("Scatter Plots:\n")
features = ['ph', 'Sulfate', 'Trihalomethanes', 'Chloramines', 'Conductivity', 'Hardness', 'Turbidity', 'Organic_carbon', 'Solids']
for i in range(0, len(features)-1, 2):
    df_imputed.plot.scatter(x=features[i], y=features[i+1])
    plt.show()

# Matrice de corrélation
print(df.iloc[:, 0:9].corr())

# Gestion des outliers pour chaque caractéristique
features = ['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']
for feature in features:
    Q1 = df_imputed[feature].quantile(0.25)
    Q3 = df_imputed[feature].quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5
    outliers = df_imputed[(df_imputed[feature] < Q1 - threshold * IQR) | (df_imputed[feature] > Q3 + threshold * IQR)]
    df_imputed.loc[outliers.index, feature] = df[feature].mean()

# Calcul des variances
print("Variance for each feature:")
for feature in features:
    variance = df_imputed[feature].var()
    print(f"{feature}: {variance}")

# Division des données en ensemble d'entraînement et de test
x = df_imputed.iloc[:, :-1]
y = df_imputed["Potability"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(x_train.shape, y_train.shape)

# Création et entraînement du modèle Decision Tree
treeClassifier = DecisionTreeClassifier()
treeClassifier.fit(x_train, y_train)

y_test_pred = treeClassifier.predict(x_test)
# Affichage des propriétés de l'arbre
print("Number of leaves:", treeClassifier.get_n_leaves())
print("Tree depth:", treeClassifier.get_depth())
plt.figure(figsize=(20, 10))
plot_tree(treeClassifier, filled=True)
plt.show()
# Calculate the accuracy of the model on the test set
accuracy_test = accuracy_score(y_test, y_test_pred)

print("Accuracy on the test set of decesion tree:", accuracy_test)
# Initialize and train a logistic regression model
model = LogisticRegression()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Calculate the accuracy of the model on the test set
accuracy_test = accuracy_score(y_test, y_pred)
print("Accuracy on the test set of LogisticRegression :", accuracy_test)
# Initialize and train a SVM model
rs = 123
model = SVC()
model.fit(x_train, y_train.values.ravel())
preds = model.predict(x_test)
# Calculate the accuracy of the model on the test set
accuracy_test = accuracy_score(y_test, preds)
print("Accuracy on the test set of SVM :", accuracy_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
x = df_imputed.iloc[:, :-1]
y = df_imputed["Potability"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(x_train.shape, y_train.shape)
# Decision Tree
param_grid_tree = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_tree = GridSearchCV(DecisionTreeClassifier(), param_grid_tree, cv=5, scoring='accuracy')
grid_tree.fit(x_train, y_train)
best_tree = grid_tree.best_estimator_
print("Best parameters for Decision Tree:", grid_tree.best_params_)
print("Best Decision Tree Accuracy:", grid_tree.best_score_)

# Logistic Regression
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'saga']
}
grid_lr = GridSearchCV(LogisticRegression(max_iter=10000), param_grid_lr, cv=5, scoring='accuracy')
grid_lr.fit(x_train, y_train)
best_lr = grid_lr.best_estimator_
print("Best parameters for Logistic Regression:", grid_lr.best_params_)
print("Best Logistic Regression Accuracy:", grid_lr.best_score_)

```

```python
# SVM
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
grid_svm = GridSearchCV(SVC(), param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(x_train, y_train)
best_svm = grid_svm.best_estimator_
print("Best parameters for SVM:", grid_svm.best_params_)
print("Best SVM Accuracy:", grid_svm.best_score_)
from sklearn.ensemble import VotingClassifier

voting_clf = VotingClassifier(estimators=[
    ('dt', best_tree),
    ('lr', best_lr),
    ('svm', best_svm)
], voting='hard')

voting_clf.fit(x_train, y_train)
voting_pred = voting_clf.predict(x_test)
voting_accuracy = accuracy_score(y_test, voting_pred)
print("Voting Classifier Accuracy:", voting_accuracy)
```

//1. Objectif principal de l'analyse
L'objectif principal de cette analyse est de construire un modèle de classification capable de prédire la potabilité de l'eau en fonction de diverses caractéristiques chimiques et physiques. Ce modèle peut être utilisé par des entreprises ou des organismes de santé pour évaluer la qualité de l'eau potable, ce qui est crucial pour la santé publique. En plus de la prédiction, cette analyse permet d'identifier les principales caractéristiques influençant la potabilité de l'eau.

2. Description de l'ensemble de données
L'ensemble de données utilisé contient des informations sur différentes caractéristiques de l'eau, telles que le pH, la dureté, les solides dissous, la concentration en chloramines, etc. La variable cible est la potabilité, indiquant si l'eau est potable (1) ou non (0). L'analyse se concentre sur le nettoyage des données, la gestion des valeurs manquantes et l'identification des valeurs aberrantes pour assurer une modélisation efficace.

3. Exploration des données et nettoyage
L'exploration des données a révélé la présence de valeurs manquantes dans certaines colonnes. Ces valeurs manquantes ont été supprimées pour simplifier l'analyse. Des visualisations telles que des boxplots et des histogrammes ont été utilisées pour comprendre la distribution des données et identifier des valeurs aberrantes. Une gestion des outliers a été effectuée en utilisant la méthode de l'écart interquartile (IQR), en remplaçant les valeurs aberrantes par la moyenne des caractéristiques concernées.

4. Formation des modèles de classification
Trois modèles de classification ont été formés :

Decision Tree : Ce modèle a montré une précision de base de 62.78% sur l'ensemble de test.
Logistic Regression : Modèle de référence avec une précision de 62.28%.
Support Vector Machine (SVM) : Ce modèle a également obtenu une précision de 62.78%, similaire à celle du Decision Tree.
Après avoir effectué une recherche de grille (GridSearchCV) pour optimiser les hyperparamètres, les performances des modèles se sont améliorées. Les meilleurs modèles ont été combinés dans un Voting Classifier, qui a abouti à une précision finale de 65.12%.

5. Recommandation du modèle final
Le modèle Voting Classifier est recommandé comme modèle final car il combine les forces des trois modèles individuels (Decision Tree, Logistic Regression, SVM). En combinant les prédictions, il améliore légèrement la précision et offre une robustesse accrue par rapport aux modèles individuels.

6. Conclusions et insights
L'analyse a montré que certaines caractéristiques comme le pH, les solides dissous, et les trihalométhanes influencent significativement la potabilité de l'eau. Le Voting Classifier a permis d'améliorer la précision globale de la classification, ce qui est essentiel pour prendre des décisions fiables concernant la qualité de l'eau.
